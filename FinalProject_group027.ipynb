## Dataset #2 (Unemployment in America Per US State)

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the data
df_unemployment = pd.read_csv('Unemployment in America Per US State.csv')


# Inspect the data
print(df_unemployment.head())
print(df_unemployment.info())


# Remove commas and convert columns to numeric
columns_to_convert = [
    'Total Civilian Non-Institutional Population in State/Area',
    'Total Civilian Labor Force in State/Area',
    'Total Employment in State/Area',
    'Total Unemployment in State/Area'
]


for column in columns_to_convert:
    df_unemployment[column] = df_unemployment[column].str.replace(',', '').astype(int)


# Verify changes
print(df_unemployment.info())


# Check for missing values
print(df_unemployment.isnull().sum())


# Remove duplicates
df_unemployment.drop_duplicates(inplace=True)


# Standardize text
df_unemployment['State/Area'] = df_unemployment['State/Area'].str.strip().str.upper()


# Rename columns
df_unemployment.rename(columns={
    'FIPS Code': 'FIPS_Code',
    'State/Area': 'State_Area',
    'Year': 'Year',
    'Month': 'Month',
    'Total Civilian Non-Institutional Population in State/Area': 'Total_Civilian_Population',
    'Total Civilian Labor Force in State/Area': 'Labor_Force',
    'Percent (%) of State/Area\'s Population': 'Percent_Population',
    'Total Employment in State/Area': 'Total_Employment',
    'Percent (%) of Labor Force Employed in State/Area': 'Percent_Employed',
    'Total Unemployment in State/Area': 'Total_Unemployment',
    'Percent (%) of Labor Force Unemployed in State/Area': 'Percent_Unemployed'
}, inplace=True)


# Filter data for the years 2017-2022
df_filtered1 = df_unemployment[(df_unemployment['Year'] >= 2017) & (df_unemployment['Year'] <= 2022)]


# Group by State/Area and Year, then calculate the mean unemployment rate
df_grouped = df_filtered1.groupby(['State_Area', 'Year'])['Percent_Unemployed'].mean().reset_index()


# Set the plot size
plt.figure(figsize=(14, 10))


# Create a lineplot for each state's unemployment rate over the years
sns.lineplot(data=df_grouped, x='Year', y='Percent_Unemployed', hue='State_Area', palette='tab20', legend=None)


# Add titles and labels
plt.title('Change in Unemployment Rate for Each State (2017-2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.xticks(range(2017, 2023))


# Show the plot
plt.show()
```

This graph shows that the unemployment rate starts to increase in 2019 until 2020 when it is at its highest. This coincides with the beginning and height of the COVID-19 Pandemic. 2021 marks a significant dip in the unemployment rate, which continues to lower as the year approaches 2022.

# Limitations

There are numerous variables that have the potential to affect the economy, poverty rates, as well as unemployment rates in different ways. These variables include, but are not limited to, economic cycles, demographic changes, and external factors such as global economic conditions. As a result, there were limitations within our analysis.

One limitation is related to the Poverty Status data from the U.S. Census. It is important to note that the data was not actual values reported, but rather predictions calculated by the U.S. Census Bureau. These predictions are based on 5-year estimates, which generally do have smaller margins of error than 1-year estimates. However, there are a number of factors that affect the accuracy of these “best guess” estimates.

For example, their methods may be outdated, which could lead to data inaccuracies. It is important to note that there has been historical undercounts of young children which are correlated with inaccurate measurement of certain racial and ethnic groups. These undercounts can result in misrepresentation of data and are not truly representative of poverty status within the United States.
An additional limitation lies in the fact that some of our data’s numbers do not account for population increases. For example, the gun violence dataset is not adjusted to the country’s growing population, which is in general, a confounding variable when analyzing increases in deaths.

If we were to conduct this research project again, we would use a dataset that more accurately reflects the poverty status of the population instead of relying on a dataset of predictions. This could be finding actual reported data, if available, or one that has more frequent data collection in order to improve the accuracy of the data. We would also find gun violence data that shows the gun death rate per amount of people (i.e. per 100,000 people). By addressing these limitations, we aim to enhance the overall reliability and validity of our findings.
