## Dataset #2 (Unemployment in America Per US State)

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the data
df_unemployment = pd.read_csv('Unemployment in America Per US State.csv')


# Inspect the data
print(df_unemployment.head())
print(df_unemployment.info())


# Remove commas and convert columns to numeric
columns_to_convert = [
    'Total Civilian Non-Institutional Population in State/Area',
    'Total Civilian Labor Force in State/Area',
    'Total Employment in State/Area',
    'Total Unemployment in State/Area'
]


for column in columns_to_convert:
    df_unemployment[column] = df_unemployment[column].str.replace(',', '').astype(int)


# Verify changes
print(df_unemployment.info())


# Check for missing values
print(df_unemployment.isnull().sum())


# Remove duplicates
df_unemployment.drop_duplicates(inplace=True)


# Standardize text
df_unemployment['State/Area'] = df_unemployment['State/Area'].str.strip().str.upper()


# Rename columns
df_unemployment.rename(columns={
    'FIPS Code': 'FIPS_Code',
    'State/Area': 'State_Area',
    'Year': 'Year',
    'Month': 'Month',
    'Total Civilian Non-Institutional Population in State/Area': 'Total_Civilian_Population',
    'Total Civilian Labor Force in State/Area': 'Labor_Force',
    'Percent (%) of State/Area\'s Population': 'Percent_Population',
    'Total Employment in State/Area': 'Total_Employment',
    'Percent (%) of Labor Force Employed in State/Area': 'Percent_Employed',
    'Total Unemployment in State/Area': 'Total_Unemployment',
    'Percent (%) of Labor Force Unemployed in State/Area': 'Percent_Unemployed'
}, inplace=True)


# Filter data for the years 2017-2022
df_filtered1 = df_unemployment[(df_unemployment['Year'] >= 2017) & (df_unemployment['Year'] <= 2022)]


# Group by State/Area and Year, then calculate the mean unemployment rate
df_grouped = df_filtered1.groupby(['State_Area', 'Year'])['Percent_Unemployed'].mean().reset_index()


# Set the plot size
plt.figure(figsize=(14, 10))


# Create a lineplot for each state's unemployment rate over the years
sns.lineplot(data=df_grouped, x='Year', y='Percent_Unemployed', hue='State_Area', palette='tab20', legend=None)


# Add titles and labels
plt.title('Change in Unemployment Rate for Each State (2017-2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.xticks(range(2017, 2023))


# Show the plot
plt.show()
```

### Section 2 of EDA - Poverty Status and Unemployment Rate

In this exploratory data analysis (EDA), we investigate the relationship between poverty rates and unemployment rates across different states in the United States from 2017 to 2022. The process involves data preparation, merging datasets, and performing various analyses to understand the correlation between these socioeconomic factors.
Data Preparation and Cleaning

    Concatenating Poverty Data:
        We start by combining multiple DataFrames containing poverty status data from different years (2017 to 2022) into a single DataFrame (df_poverty). This helps in creating a consolidated view of poverty data over the years.
        We reset the index to bring the Year column into the main DataFrame, making it easier to work with.

    Renaming Columns:
        The columns are renamed to make them more descriptive and easier to understand. For example, columns related to Alabama's poverty data are renamed to distinguish between the total population, those below the poverty level, and the percentage below the poverty level.

    Filtering and Reshaping Data:
        We filter the DataFrame to retain only the relevant metrics: 'Population for whom poverty status is determined' and 'UNRELATED INDIVIDUALS FOR WHOM POVERTY STATUS IS DETERMINED'.
        The DataFrame is then reshaped to a long format using the melt function, which makes it easier to handle and analyze.

    Extracting State and Metric:
        We extract the state names and poverty metrics from the State_Metric column by splitting it into two separate columns: State and Poverty_Metric.
        State names are standardized to uppercase to ensure consistency.

    Data Type Conversion:
        The percentage values in the poverty data are converted to numeric by removing the '%' symbol. This allows for numerical analysis and plotting.

    Pivoting Data:
        The DataFrame is pivoted to create separate columns for each poverty metric, indexed by Year and State. This transformation helps in merging the data with unemployment data later.

    Loading and Cleaning Unemployment Data:
        We load the unemployment data from a CSV file and clean it by converting relevant columns to integer type after removing commas. This ensures the data is in a usable format for analysis.
        Columns are renamed for clarity, and state names are standardized to uppercase.
        The DataFrame is filtered to retain data from 2017 to 2022.

Merging and Analyzing Data

    Merging Poverty and Unemployment Data:
        We merge the cleaned poverty and unemployment data on the common attributes of State and Year. This creates a comprehensive dataset that includes both poverty and unemployment information for each state and year.
        Unnecessary columns are dropped to keep the dataset focused and manageable.

    Summary Statistics and Correlation:
        We calculate summary statistics for the unemployment and poverty rates to get an overview of the data distribution.
        A correlation matrix is computed to quantify the relationship between unemployment rates and poverty rates. This matrix is visualized using a heatmap to easily interpret the correlations.

    State-wise Analysis:
        The data is grouped by state, and the mean values of unemployment and poverty rates are calculated. This provides insights into how these rates vary across different states.
        A scatter plot is created to visualize the relationship between average unemployment rates and average poverty rates for each state. This helps identify any patterns or outliers.

    Yearly Trends Analysis:
        The data is grouped by year, and the mean values of unemployment and poverty rates are calculated to observe trends over time.
        A dual-axis plot is created to show the trends of poverty and unemployment rates from 2017 to 2022. This visualization helps understand how these rates have changed over the years and whether they follow similar patterns.

This EDA provides a detailed examination of the relationship between poverty and unemployment rates across different states and over multiple years. By combining, cleaning, and visualizing the data, we gain valuable insights into the socioeconomic dynamics at play, which can inform policy decisions and further research.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of poverty data frames
poverty_dfs = [povertystatus_2017, povertystatus_2018, povertystatus_2019, povertystatus_2020, povertystatus_2021, povertystatus_2022]

# Concatenate all data frames
df_poverty = pd.concat(poverty_dfs, keys=['2017', '2018', '2019', '2020', '2021', '2022'], names=['Year'])
df_poverty.reset_index(level=0, inplace=True)

# Rename columns
df_poverty.rename(columns={
    'Label (Grouping)': 'Metric',
    'Alabama!!Total!!Estimate': 'Alabama_Total',
    'Alabama!!Below poverty level!!Estimate': 'Alabama_Below_Poverty',
    'Alabama!!Percent below poverty level!!Estimate': 'Alabama_Poverty_Percent',
    # Repeat for other states
}, inplace=True)

# Filter to keep only relevant metrics
df_poverty_filtered = df_poverty[df_poverty['Metric'].isin(['Population for whom poverty status is determined', 'UNRELATED INDIVIDUALS FOR WHOM POVERTY STATUS IS DETERMINED'])]

# Melt DataFrame to long format
df_poverty_long = pd.melt(df_poverty_filtered, id_vars=['Year', 'Metric'], var_name='State_Metric', value_name='Value')

# Adjust the split operation based on the structure of State_Metric
# Example: If State_Metric is in the form 'State!!PovertyMetric'
df_poverty_long[['State', 'Poverty_Metric']] = df_poverty_long['State_Metric'].str.extract(r'^(.*?)!!(.*?!!Estimate)$')

# Verify the extraction
print("DataFrame after extracting State and Poverty_Metric:")
print(df_poverty_long.head())

# Standardize state names
df_poverty_long['State'] = df_poverty_long['State'].str.strip().str.upper()

# Convert 'Percent below poverty level!!Estimate' to numeric
df_poverty_long.loc[df_poverty_long['Poverty_Metric'] == 'Percent below poverty level!!Estimate', 'Value'] = (
    df_poverty_long.loc[df_poverty_long['Poverty_Metric'] == 'Percent below poverty level!!Estimate', 'Value']
    .str.replace('%', '')
    .astype(float)
)

# Pivot to get separate columns for each metric
df_poverty_pivot = df_poverty_long.pivot_table(index=['Year', 'State'], columns='Poverty_Metric', values='Value', aggfunc='first').reset_index()

# Verify the transformation
print("Poverty data after pivot:")
print(df_poverty_pivot.head())

# Ensure the 'Year' column in both DataFrames is of the same type (int)
df_poverty_pivot['Year'] = df_poverty_pivot['Year'].astype(int)

# Load the unemployment data
file_path_unemployment = 'Unemployment in America Per US State.csv'
df_unemployment = pd.read_csv(file_path_unemployment)

# Clean the unemployment data (similar to previous instructions)
columns_to_convert = [
    'Total Civilian Non-Institutional Population in State/Area',
    'Total Civilian Labor Force in State/Area',
    'Total Employment in State/Area',
    'Total Unemployment in State/Area'
]

for column in columns_to_convert:
    df_unemployment[column] = df_unemployment[column].str.replace(',', '').astype(int)

df_unemployment['State/Area'] = df_unemployment['State/Area'].str.strip().str.upper()

df_unemployment.rename(columns={
    'FIPS Code': 'FIPS_Code',
    'State/Area': 'State_Area',
    'Year': 'Year',
    'Month': 'Month',
    'Total Civilian Non-Institutional Population in State/Area': 'Total_Civilian_Population',
    'Total Civilian Labor Force in State/Area': 'Labor_Force',
    'Percent (%) of State/Area\'s Population': 'Percent_Population',
    'Total Employment in State/Area': 'Total_Employment',
    'Percent (%) of Labor Force Employed in State/Area': 'Percent_Employed',
    'Total Unemployment in State/Area': 'Total_Unemployment',
    'Percent (%) of Labor Force Unemployed in State/Area': 'Percent_Unemployed'
}, inplace=True)

# Ensure the 'Year' column in the unemployment DataFrame is of type int
df_unemployment['Year'] = df_unemployment['Year'].astype(int)

# Filter unemployment data for the relevant years
df_filtered = df_unemployment[(df_unemployment['Year'] >= 2017) & (df_unemployment['Year'] <= 2022)]

# Standardize state names in unemployment data
df_filtered['State_Area'] = df_filtered['State_Area'].str.strip().str.upper()

# Merge poverty and unemployment data
df_merged = pd.merge(
    df_poverty_pivot,
    df_filtered[['State_Area', 'Year', 'Percent_Unemployed']],
    left_on=['State', 'Year'],
    right_on=['State_Area', 'Year'],
    how='inner'
)

# Drop unnecessary columns
df_merged.drop(columns=['State_Area'], inplace=True)

# Verify merge
print("Merged DataFrame:")
print(df_merged.head())

# Summary statistics for unemployment and poverty rates
summary_stats = df_merged[['Percent_Unemployed', 'Percent below poverty level!!Estimate']].describe()
print("Summary Statistics:")
print(summary_stats)

# Compute correlation matrix
correlation_matrix = df_merged[['Percent_Unemployed', 'Percent below poverty level!!Estimate']].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Poverty and Unemployment Rates')
plt.show()

# Group by state and calculate mean values
state_analysis = df_merged.groupby('State').agg({
    'Percent_Unemployed': 'mean',
    'Percent below poverty level!!Estimate': 'mean'
}).reset_index()

# Plot state-wise analysis
plt.figure(figsize=(14, 7))
sns.scatterplot(data=state_analysis, x='Percent_Unemployed', y='Percent below poverty level!!Estimate', hue='State', legend=False)
plt.title('State-wise Poverty vs. Unemployment Rate (2017-2022)')
plt.xlabel('Average Unemployment Rate (%)')
plt.ylabel('Average Poverty Rate (%)')
plt.show()

# Group by year and calculate mean values
yearly_trends = df_merged.groupby('Year').agg({
    'Percent_Unemployed': 'mean',
    'Percent below poverty level!!Estimate': 'mean'
}).reset_index()

# Plot yearly trends
fig, ax1 = plt.subplots(figsize=(14, 7))

ax2 = ax1.twinx()
ax1.plot(yearly_trends['Year'], yearly_trends['Percent below poverty level!!Estimate'], 'g-')
ax2.plot(yearly_trends['Year'], yearly_trends['Percent_Unemployed'], 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Average Poverty Rate (%)', color='g')
ax2.set_ylabel('Average Unemployment Rate (%)', color='b')

plt.title('Yearly Trends of Poverty and Unemployment Rates (2017-2022)')
plt.show()
```
